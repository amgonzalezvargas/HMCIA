```python
#------------------------------
def is_diagonal(matrix):
    rows, cols = matrix.shape
    # Verificar si la matriz es cuadrada
    if rows != cols:
        return False
    for i in range(rows):
        for j in range(cols):
            if i != j and matrix[i, j] != 0:
                return False
    return True

#------------------------------

# Ejemplo de uso:
A= np.array([[-1, 0, 0],
             [ 0, 2, 0],
             [ 0, 1, 1]])
B= np.array([[ 1, 0, 0],
             [ 0, 0, 0],
             [ 0, 0, 2]])
C= np.array([[ 1, 0, 0],
             [ 0, 2, 0]])


print(is_diagonal(A))  # Debería imprimir True
print(is_diagonal(B))  # Debería imprimir True
print(is_diagonal(C))  # Debería imprimir True
```

```python
start=-10; end=10; step=0.1
x = np.arange(start, end + step, step)
y = np.arange(start, end + step, step)

# Create a meshgrid of x and y
X, Y = np.meshgrid(x, y)
FXY = (np.sqrt(X**2 + Y**2 - 9)) / X

# Create a single figure with two subplots in one row
fig = plt.figure(figsize=(12, 6))

# First subplot (3D plot)
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, FXY)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('FXY')

# Second subplot (2D heatmap)
ax2 = fig.add_subplot(122)
cmap = "viridis"
heatmap = ax2.pcolormesh(X, Y, FXY, cmap=cmap, shading='auto')
cbar = fig.colorbar(heatmap, ax=ax2)
cbar.set_label('FXY')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.tight_layout()
plt.show()

print(FXY)
```

```python
# Define the symbolic variable 'x' using SymPy
x = sym.symbols('x')

# Calculate the derivative of sqrt(x) with respect to 'x'
expr = sym.diff(sym.sqrt(x), x)

# Print the derivative expression
print(expr)

# Evaluate the derivative at 'x = 1' and print the result
print(expr.evalf(subs={x: 2}))
print(1/(2*np.sqrt(2)))
```

#Taller de Calculo de Probabilidades

##1
image.png

```python
# Importar la librería itertools para generar todas las posibles combinaciones
import itertools
# Definir los resultados posibles de lanzar una moneda imparcial
resultados_moneda = ['Cara', 'Cruz']
# Generar todas las posibles combinaciones de resultados para dos monedas
combinaciones = list(itertools.product(resultados_moneda, repeat=2))
# Calcular la probabilidad de cada evento
total_combinaciones = len(combinaciones)
print(combinaciones)
```

```python
# Evento A: al menos una cara
event_A = [combo for combo in combinaciones if 'Cara' in combo]
prob_A = len(event_A) / total_combinaciones

# Evento B: al menos una cruz
event_B = [combo for combo in combinaciones if 'Cruz' in combo]
prob_B = len(event_B) / total_combinaciones

# Evento A ∩ B: al menos una cara y al menos una cruz
event_A_and_B = [combo for combo in combinaciones if 'Cara' in combo and 'Cruz' in combo]
prob_A_and_B = len(event_A_and_B) / total_combinaciones

# Evento A ∪ B: al menos una cara o al menos una cruz
event_A_or_B = [combo for combo in combinaciones if 'Cara' in combo or 'Cruz' in combo]
prob_A_or_B = len(event_A_or_B) / total_combinaciones

# Evento A^c: ninguna cara
event_A_complement = [combo for combo in combinaciones if 'Cara' not in combo]
prob_A_complement = len(event_A_complement) / total_combinaciones

# Imprimir las probabilidades
print(f"Probabilidad de A: {prob_A:.2f}")
print(f"Probabilidad de B: {prob_B:.2f}")
print(f"Probabilidad de A ∩ B: {prob_A_and_B:.2f}")
print(f"Probabilidad de A ∪ B: {prob_A_or_B:.2f}")
print(f"Probabilidad de A^c: {prob_A_complement:.2f}")
```

```python
import random

# Función para lanzar una moneda (Cara o Cruz)
def lanzar_moneda():
    return random.choice(['Cara', 'Cruz'])

# Simular el lanzamiento de dos monedas
num_lanzamientos = 10000
resultados = [(lanzar_moneda(), lanzar_moneda()) for _ in range(num_lanzamientos)]
print(len(resultados))
print(resultados)
```

```python
# Contar la ocurrencia de eventos A y B
contador_A = 0
contador_B = 0
contador_A_dado_B = 0
contador_A_dado_no_B = 0
contador_B_dado_A = 0
contador_B_dado_no_A = 0

for moneda1, moneda2 in resultados:
    if moneda1 == 'Cara':
        contador_A += 1
    if moneda2 == 'Cruz':
        contador_B += 1
    if moneda1 == 'Cara' and moneda2 == 'Cruz':
        contador_A_dado_B += 1
    if moneda1 == 'Cara' and moneda2 != 'Cruz':
        contador_A_dado_no_B += 1
    if moneda2 == 'Cruz' and moneda1 == 'Cara':
        contador_B_dado_A += 1
    if moneda2 == 'Cruz' and moneda1 != 'Cara':
        contador_B_dado_no_A += 1

# Calcular probabilidades condicionales
prob_A_dado_B = contador_A_dado_B / contador_B
prob_A_dado_no_B = contador_A_dado_no_B / (num_lanzamientos - contador_B)
prob_B_dado_A = contador_B_dado_A / contador_A
prob_B_dado_no_A = contador_B_dado_no_A / (num_lanzamientos - contador_A)

# Imprimir las probabilidades condicionales
print(f"P(A|B) = {prob_A_dado_B:.2f}")
print(f"P(A|¬B) = {prob_A_dado_no_B:.2f}")
print(f"P(B|A) = {prob_B_dado_A:.2f}")
print(f"P(B|¬A) = {prob_B_dado_no_A:.2f}")
```

```python
P_H = 0.60  # Probabilidad de que le guste el helado
P_T = 0.36  # Probabilidad de que le guste la torta
P_T_given_H = 0.40  # Probabilidad de que le guste la torta dado que le gusta el helado
P_H_given_T = (P_T_given_H * P_H) / P_T
print(f"La probabilidad de que a un alumno le guste el helado dado que le gusta la torta es: {P_H_given_T:.4f} o {P_H_given_T*100:.2f}%")
```

image.png

```python
# Probabilidades iniciales
P_I = 0.20
P_E = 0.20
P_D_given_I = 0.75
P_D_given_E = 0.50
P_D_given_not_I_E = 0.20  # No es ni ingeniero ni economista

# Calculamos la probabilidad de no ser ni ingeniero ni economista
P_not_I_E = 1 - P_I - P_E

# Usando el principio de probabilidad total para encontrar P(D)
P_D = P_D_given_I * P_I + P_D_given_E * P_E + P_D_given_not_I_E * P_not_I_E

# Usamos el Teorema de Bayes para encontrar P(I|D)
P_I_given_D = (P_D_given_I * P_I) / P_D

print(f"La probabilidad de que un empleado directivo elegido al azar sea ingeniero es: {P_I_given_D:.4f} o {P_I_given_D*100:.2f}%")
```

```python
# Probabilidades dadas
P_I = 0.1
P_A_given_I = 0.97
P_A_given_not_I = 0.02

# Probabilidad complementaria de que ocurra un incidente
P_not_I = 1 - P_I

# Calculamos la probabilidad de que suene la alarma usando la ley de probabilidad total
P_A = P_A_given_I * P_I + P_A_given_not_I * P_not_I

# Usamos el Teorema de Bayes para encontrar P(I^c|A)
P_not_I_given_A = (P_A_given_not_I * P_not_I) / P_A

print(f"La probabilidad de que no haya habido un incidente dado que sonó la alarma es: {P_not_I_given_A:.4f} o {P_not_I_given_A*100:.2f}%")
```
